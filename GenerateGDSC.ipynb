```python
import numpy as np
import pandas as pd
import math
```

```python
# Come from 
# https://www.cancerrxgene.org/gdsc1000/GDSC1000_WebResources//Data/preprocessed/Cell_line_RMA_proc_basalExp.txt.zip
data = pd.read_csv("Cell_line_RMA_proc_basalExp.txt",sep="\t",)
```

```python
select_drug = 'Cisplatin'
```

```python
data_nona=data.dropna()
```

```python
g_smbols = data_nona.GENE_SYMBOLS
```

```python
g_smbols
```

```python
data_nona=data_nona.drop(["GENE_title","GENE_SYMBOLS"],axis=1)
```

```python
data_nona = data_nona.T
```

```python
data_nona.columns = g_smbols
```

```python
data_nona
```

```python
# Comes from 
#ftp://ftp.sanger.ac.uk/pub/project/cancerrxgene/releases/current_release/GDSC2_fitted_dose_response_25Feb20.xlsx
label = pd.read_excel("GDSC2_fitted_dose_response_25Feb20.xlsx")
```

```python
label_info = label.loc[:,["COSMIC_ID","CELL_LINE_NAME","AUC","DRUG_NAME"]]
```

```python
selected_drug_sens = label_info.loc[label_info.DRUG_NAME==select_drug]
```

```python
transform_CID = [ "DATA."+str(i) for i in selected_drug_sens.COSMIC_ID] 
```

```python
select_cells = data_nona.loc[transform_CID,:]
```

```python
selected_drug_sens.to_csv("GDSC2_label_cisplatin.csv")
```

```python
select_cells.to_csv("GDSC2_expression_cisplatin.csv")
```
